# Unit Tests

In the previous notebook you became familiar with the assert statement. Assert statements are a way of quickly checking if a certain snippet of code returns the expected output. 

Unit testing, on the other hand, is a structured framework for testing codebases in a collaborative environment (any developer job). A rule of thumb is that assert statements will help you here and now, while unit testing is something you take the time to implement in order to help others using or contributing to your code later. 

Here, we will only dip into the surface of unit testing.
